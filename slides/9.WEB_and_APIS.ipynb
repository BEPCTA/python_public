```python
from requests import get, post
from bs4 import BeautifulSoup
G_URL = 'https://www.google.ru/search'

def google(query):
    req = get(G_URL, params = {'q' : query})    
    assert req.status_code == 200, 'request failed'
    soup = BeautifulSoup(req.text, "lxml")
    
    for i, li in enumerate(soup.findAll('h3', attrs={'class' : 'r'})):
        link = li.find('a')
        if link:
            link['href'] = 'http://www.google.com' +  link['href']
            link = '<i>{}.</i>  {}'.format(i, link)
            display(HTML(str(link)))

google('Тест!')
```

### Непубличные API

Любые WEB-приложения для обновления данных внутри себя ходят в какие-то внутренние API своей платформы.
При некотором старании можно научиться пользоваться этими API в вашем коде.

```python
from requests import post
from json import loads, dumps

def get_doublegis(source, dest):
    url = 'https://catalog.api.2gis.ru/carrouting/4.0.0/moscow?key=rutnpt3272'
    
    payload = { "locale":"en",  "point_a_name":"Start", "point_b_name":"Finish", "points" : [{
                "type":"pedo",
                "x":source[1],
                "y":source[0]
            }, {
                "type":"pedo",
                "x":dest[1],
                "y":dest[0]
            }
        ], "type":"online", "region_id":"32"
    }
 

    data  = post(url, data=dumps(payload))
    route = loads(data.text)['result'][0]
    length = route['total_distance']
    eta    = route['total_duration']

    return length, eta / 60
```

```python
source, dest = (55.734242, 37.587424), (55.798485, 37.39912)
print(get_doublegis(source, dest))
```
